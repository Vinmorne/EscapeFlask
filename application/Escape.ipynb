Vous foncez dans le couloir de gauche, jusqu'au bout. C'est une impasse : il n'y a que 6 leviers étiquetés A, E, I, O, U et Y.

Derrière les leviers, il y a un muret qui isole du reste du couloir 4 petits récipients comportant chacun 5 perles rouges.

En actionnant les leviers les uns après les autres vous vous apercevez qu'ils font basculer certains récipients de façon à :
- levier A : passe une perle de 1 à 2
- levier E : passe une perle de 2 à 3
- levier I : passe une perle de 3 à 4
- levier O : passe une perle de 4 à 1
- levier U : passe (jusqu'à) deux perles de 4 à 2
- levier Y : passe (jusqu'à) quatre perles de 3 à 1

Evidemment lorsqu'un récipient est vide, il ne peut plus passer une perle ailleurs...

Quand les billes sont exactement en bon nombre dans les quatre récipients, le mur du fond s'avère être amovible. Un mécanisme lui permet de coulisser vers le plafond.

C'est l'entrée d'une grand pièce, dont les murs sont tapissés de casiers, avec des identifiants inscrits dessus... des coffres !

Les identifiants sont tous du même format : deux lettres en majuscules, puis 3 chiffres et enfin un caractère spécial, toujours parmi ceux-ci : 
```
§$*!?+=&@#%
```

Quels sont les coffres que vous allez chercher à ouvrir ?



````python
# génération codes
import random
spe = "§$*!?+=&@#%"
let = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
chi = "0123456789"

coffres = []

def myHash(text:str):
  hash=0
  for ch in text:
    hash = ( hash*281  ^ ord(ch)*997) & 0xFFFFFFFFFFFF
  return hash

def gen_coffre():
  matricule = random.choice(let)
  matricule += random.choice(let)
  for i in range(3): matricule += random.choice(chi)
  matricule += random.choice(spe)
  return matricule

list_coffres = list()
for i in range(15):
  mat = gen_coffre()
  list_coffres.append((mat, myHash(mat)))

for m, h in list_coffres:
  print(m)
print([m for m, h in list_coffres])
print([h for m, h in list_coffres])


print(myHash("AB345+"))

# décodage
hashed = [154994408424555, 227394243688122, 228506018615843, 125424345289606, 64010149558732, 94133629483623, 161308262970238, 137723878758118, 254554304223548, 49000078573088, 53644746289687, 218516147973373, 176901426640253, 47704786186869, 253887343040961]

coffres = ['YK477!', 'PL846$', 'IZ855$', 'GQ256!', 'FW314+', 'GL366+', 'HQ409*', 'OP214&', 'IV420%', 'ZZ830+', 'NY324!', 'YE240#', 'FQ292?', 'GC535&', 'TO394@']

def in_hashed(mat):
  h = myHash(mat)
  return h in hashed

reponse = list()

for c1 in let:
  for c2 in let:
    for c3 in chi:
      for c4 in chi:
        for c5 in chi:
          for c6 in spe:
            mat = c1+c2+c3+c4+c5+c6
            if in_hashed(mat): reponse.append(mat)
print(reponse)
  
````
